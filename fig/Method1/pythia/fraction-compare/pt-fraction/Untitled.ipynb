```python
from ROOT import *
import numpy as np
import uncertainties as unc # propagate uncertainties
from uncertainties import unumpy as unp # array operations for type ufloat
import os


        
def myText(x,y,text,color=1):
    l = TLatex()
    l.SetTextSize(0.025)
    l.SetNDC()
    l.SetTextColor(color)
    l.DrawLatex(x,y,text)
    pass

var = "pt"


fpythia = TFile("/eos/user/r/rqian/dijet-mono-result/pythia/dijet_pythia_mono.root")
fsherpa = TFile("/eos/user/r/rqian/dijet-mono-result/sherpa/dijet_sherpa.root")
fherdipo = TFile("/eos/user/r/rqian/dijet-mono-result/herdipo/dijet_herdipo.root")
fherang = TFile("/eos/user/r/rqian/dijet-mono-result/herang/dijet_herang.root")
fsherpa_lund = TFile("/eos/user/r/rqian/dijet-mono-result/sherpa_lund/dijet_sherpa_lund.root")
fpowpyt = TFile("/eos/user/r/rqian/dijet-mono-result/powpyt/dijet_powpyt.root")

f_mc_generators = [fpythia,fsherpa,fsherpa_lund,fherang,fherdipo,fpowpyt]
mc_generators_name = ["pythia","sherpa","sherpa-lund","herang","herdipo","powheg-pythia"]



particle_array = ["Quark","Gluon","B Quark","C Quark"]
bin = np.array([500,600,800,1000,1200,1500,2000])

def unc_array_err(hist,err):
    value = np.zeros(hist.GetNbinsX())
    error = np.zeros(hist.GetNbinsX())
    for j in range(1,hist.GetNbinsX()+1):
        value[j-1] = hist.GetBinContent(j)
        error[j-1] = np.sqrt(err.GetBinContent(j))
    result = unp.uarray(value,error)
    return(result)

def set_hist_error(hist,unc):
    for i in range(1,hist.GetNbinsX()+1):
        hist.SetBinError(i,unp.std_devs(unc[i-1]))
        


for particle in particle_array:
    higher_hist_array =  []
    lower_hist_array =  []
    for mc_index in range(len(f_mc_generators)):
        print(mc_generators_name[mc_index])
        file1 = f_mc_generators[mc_index]
        for i in range(0,6):
            print(bin[i])        
            min = str(bin[i])
            max = str(bin[i+1])
            higher_quark = file1.Get(str(bin[i])+"_SubJet_Forward_Quark_"+var)
            higher_gluon = file1.Get(str(bin[i])+"_SubJet_Forward_Gluon_"+var)
            higher_quark1 = file1.Get(str(bin[i])+"_LeadingJet_Forward_Quark_"+var)
            higher_gluon1 = file1.Get(str(bin[i])+"_LeadingJet_Forward_Gluon_"+var)
    
            higher_quark.SetName(str(bin[i])+"_SubJet_Forward_Quark_"+var+str(mc_index))
            higher_gluon.SetName(str(bin[i])+"_SubJet_Forward_Gluon_"+var+str(mc_index))
            higher_quark1.SetName(str(bin[i])+"_LeadingJet_Forward_Quark_"+var+str(mc_index))
            higher_gluon1.SetName(str(bin[i])+"_LeadingJet_Forward_Gluon_"+var+str(mc_index))
            
            higher_quark.Add(higher_quark1)
            higher_gluon.Add(higher_gluon1)
    
            lower_quark = file1.Get(str(bin[i])+"_SubJet_Central_Quark_"+var)
            lower_gluon = file1.Get(str(bin[i])+"_SubJet_Central_Gluon_"+var)
            lower_quark1 = file1.Get(str(bin[i])+"_LeadingJet_Central_Quark_"+var)
            lower_gluon1 = file1.Get(str(bin[i])+"_LeadingJet_Central_Gluon_"+var)
            
            lower_quark.SetName(str(bin[i])+"_SubJet_Central_Quark_"+var+str(mc_index))
            lower_gluon.SetName(str(bin[i])+"_SubJet_Central_Gluon_"+var+str(mc_index))
            lower_quark1.SetName(str(bin[i])+"_LeadingJet_Central_Quark_"+var+str(mc_index))
            lower_gluon1.SetName(str(bin[i])+"_LeadingJet_Central_Gluon_"+var+str(mc_index))    
            
            lower_quark.Add(lower_quark1)
            lower_gluon.Add(lower_gluon1)
            
            higher_b_quark = file1.Get(str(bin[i])+"_SubJet_Forward_B_Quark_"+var)
            higher_b_quark1 = file1.Get(str(bin[i])+"_LeadingJet_Forward_B_Quark_"+var)
            higher_b_quark.SetName(str(bin[i])+"_SubJet_Forward_B_Quark_"+var+str(mc_index))  
            higher_b_quark1.SetName(str(bin[i])+"_LeadingJet_Forward_B_Quark_"+var+str(mc_index))        
    
            higher_b_quark.Add(higher_b_quark1)
    
            lower_b_quark = file1.Get(str(bin[i])+"_LeadingJet_Central_B_Quark_"+var)
            lower_b_quark1 = file1.Get(str(bin[i])+"_SubJet_Central_B_Quark_"+var)
            
            lower_b_quark.SetName(str(bin[i])+"_SubJet_Central_B_Quark_"+var+str(mc_index))
            lower_b_quark1.SetName(str(bin[i])+"_LeadingJet_Central_B_Quark_"+var+str(mc_index))
            
            lower_b_quark.Add(lower_b_quark1)
    
            higher_c_quark = file1.Get(str(bin[i])+"_SubJet_Forward_C_Quark_"+var)
            higher_c_quark1 = file1.Get(str(bin[i])+"_LeadingJet_Forward_C_Quark_"+var)
            
            higher_c_quark.SetName(str(bin[i])+"_SubJet_Forward_C_Quark_"+var+str(mc_index))
            higher_c_quark1.SetName(str(bin[i])+"_LeadingJet_Forward_C_Quark_"+var+str(mc_index))
            
            higher_c_quark.Add(higher_c_quark1)
    
            lower_c_quark = file1.Get(str(bin[i])+"_LeadingJet_Central_C_Quark_"+var)
            lower_c_quark1 = file1.Get(str(bin[i])+"_SubJet_Central_C_Quark_"+var)

            lower_c_quark.SetName(str(bin[i])+"_SubJet_Central_C_Quark_"+var+str(mc_index)) 
            lower_c_quark1.SetName(str(bin[i])+"_LeadingJet_Central_C_Quark_"+var+str(mc_index))
            
            lower_c_quark.Add(lower_c_quark1)
            
            higher_quark_err = file1.Get(str(bin[i])+"_SubJet_Forward_Quark_"+var+"_err")
            higher_gluon_err = file1.Get(str(bin[i])+"_SubJet_Forward_Gluon_"+var+"_err")
            higher_quark1_err = file1.Get(str(bin[i])+"_LeadingJet_Forward_Quark_"+var+"_err")
            higher_gluon1_err = file1.Get(str(bin[i])+"_LeadingJet_Forward_Gluon_"+var+"_err")
            
            higher_quark_err.Add(higher_quark1_err)
            higher_gluon_err.Add(higher_gluon1_err)
    
            lower_quark_err = file1.Get(str(bin[i])+"_LeadingJet_Central_Quark_"+var+"_err")
            lower_quark1_err = file1.Get(str(bin[i])+"_SubJet_Central_Quark_"+var+"_err")
            lower_gluon_err = file1.Get(str(bin[i])+"_LeadingJet_Central_Gluon_"+var+"_err")
            lower_gluon1_err = file1.Get(str(bin[i])+"_SubJet_Central_Gluon_"+var+"_err")
                
            lower_quark_err.Add(lower_quark1_err)
            lower_gluon_err.Add(lower_gluon1_err)
    
            higher_b_quark_err = file1.Get(str(bin[i])+"_SubJet_Forward_B_Quark_"+var+"_err")
            higher_b_quark1_err = file1.Get(str(bin[i])+"_LeadingJet_Forward_B_Quark_"+var+"_err")
            
            higher_b_quark_err.Add(higher_b_quark1_err)
    
            lower_b_quark_err = file1.Get(str(bin[i])+"_LeadingJet_Central_B_Quark_"+var+"_err")
            lower_b_quark1_err = file1.Get(str(bin[i])+"_SubJet_Central_B_Quark_"+var+"_err")
                
            lower_b_quark_err.Add(lower_b_quark1_err)
    
            higher_c_quark_err = file1.Get(str(bin[i])+"_SubJet_Forward_C_Quark_"+var+"_err")
            higher_c_quark1_err = file1.Get(str(bin[i])+"_LeadingJet_Forward_C_Quark_"+var+"_err")
            
            higher_c_quark_err.Add(higher_c_quark1_err)
    
            lower_c_quark_err = file1.Get(str(bin[i])+"_LeadingJet_Central_C_Quark_"+var+"_err")
            lower_c_quark1_err = file1.Get(str(bin[i])+"_SubJet_Central_C_Quark_"+var+"_err")
                
            lower_c_quark_err.Add(lower_c_quark1_err)
            
            higher_quark_unc_loop = unc_array_err(higher_quark,higher_quark_err)
            higher_b_quark_unc_loop = unc_array_err(higher_b_quark,higher_b_quark_err)
            higher_c_quark_unc_loop = unc_array_err(higher_c_quark,higher_c_quark_err)
    
            lower_quark_unc_loop = unc_array_err(lower_quark,lower_quark_err)
            lower_b_quark_unc_loop = unc_array_err(lower_b_quark,lower_b_quark_err)
            lower_c_quark_unc_loop = unc_array_err(lower_c_quark,lower_c_quark_err)
            
            higher_gluon_unc_loop = unc_array_err(higher_gluon,higher_gluon_err)
            lower_gluon_unc_loop = unc_array_err(lower_gluon,lower_gluon_err)
    

        
            if i == 0:
                higher_b_quark_unc = higher_b_quark_unc_loop
                higher_c_quark_unc = higher_c_quark_unc_loop
                higher_quark_unc = higher_quark_unc_loop
                higher_gluon_unc = higher_gluon_unc_loop
                higher_total_unc = higher_b_quark_unc_loop + higher_c_quark_unc_loop + higher_quark_unc_loop + higher_gluon_unc_loop
            
            else:
                higher_b_quark_unc += higher_b_quark_unc_loop
                higher_c_quark_unc += higher_c_quark_unc_loop
                higher_quark_unc += higher_quark_unc_loop
                higher_gluon_unc += higher_gluon_unc_loop
                higher_total_unc += (higher_b_quark_unc_loop + higher_c_quark_unc_loop + higher_quark_unc_loop + higher_gluon_unc_loop)
            #for j in range(0,60):
            #    print(j,higher_b_quark_unc[j],higher_total_unc[j])
    
            if i == 0:
                lower_b_quark_unc = lower_b_quark_unc_loop
                lower_c_quark_unc = lower_c_quark_unc_loop
                lower_quark_unc = lower_quark_unc_loop
                lower_gluon_unc = lower_gluon_unc_loop
                lower_total_unc = lower_b_quark_unc_loop + lower_c_quark_unc_loop + lower_quark_unc_loop + lower_gluon_unc_loop
            
            else:
                lower_b_quark_unc += lower_b_quark_unc_loop
                lower_c_quark_unc += lower_c_quark_unc_loop
                lower_quark_unc += lower_quark_unc_loop
                lower_gluon_unc += lower_gluon_unc_loop
                lower_total_unc += (lower_b_quark_unc_loop + lower_c_quark_unc_loop + lower_quark_unc_loop + lower_gluon_unc_loop)
            
        for j in range(len(higher_b_quark_unc)):
            if higher_total_unc[j].nominal_value != 0:
                higher_b_quark_unc[j] = higher_b_quark_unc[j]/higher_total_unc[j]    
                higher_c_quark_unc[j] = higher_c_quark_unc[j]/higher_total_unc[j]    
                higher_quark_unc[j] = higher_quark_unc[j]/higher_total_unc[j]  
                higher_gluon_unc[j] = higher_gluon_unc[j]/higher_total_unc[j]  
        
            else:
                higher_b_quark_unc[i] = unc.ufloat(0, 0)
                higher_c_quark_unc[i] = unc.ufloat(0, 0)
                higher_quark_unc[i] = unc.ufloat(0, 0)
                higher_gluon_unc[i] = unc.ufloat(0, 0)
    
            
        for j in range(1,higher_b_quark.GetNbinsX()+1):
            higher_b_quark.SetBinContent(j,higher_b_quark_unc[j-1].nominal_value)
            higher_c_quark.SetBinContent(j,higher_c_quark_unc[j-1].nominal_value)
            higher_quark.SetBinContent(j,higher_quark_unc[j-1].nominal_value)
            higher_gluon.SetBinContent(j,higher_gluon_unc[j-1].nominal_value)
            
            
        for j in range(len(lower_b_quark_unc)):
            if lower_total_unc[j].nominal_value != 0:
                lower_b_quark_unc[j] = lower_b_quark_unc[j]/lower_total_unc[j]    
                lower_c_quark_unc[j] = lower_c_quark_unc[j]/lower_total_unc[j]    
                lower_quark_unc[j] = lower_quark_unc[j]/lower_total_unc[j]  
                lower_gluon_unc[j] = lower_gluon_unc[j]/lower_total_unc[j]  
        
            else:
                lower_b_quark_unc[i] = unc.ufloat(0, 0)
                lower_c_quark_unc[i] = unc.ufloat(0, 0)
                lower_quark_unc[i] = unc.ufloat(0, 0)
                lower_gluon_unc[i] = unc.ufloat(0, 0)
    
            
        for j in range(1,lower_b_quark.GetNbinsX()+1):
            lower_b_quark.SetBinContent(j,lower_b_quark_unc[j-1].nominal_value)
            lower_c_quark.SetBinContent(j,lower_c_quark_unc[j-1].nominal_value)
            lower_quark.SetBinContent(j,lower_quark_unc[j-1].nominal_value)
            lower_gluon.SetBinContent(j,lower_gluon_unc[j-1].nominal_value)
            
            
            
    
        set_hist_error(lower_b_quark,lower_b_quark_unc)
        set_hist_error(lower_c_quark,lower_c_quark_unc)
        set_hist_error(lower_quark,lower_quark_unc)
        set_hist_error(lower_gluon,lower_gluon_unc)
            
    
        set_hist_error(higher_b_quark,higher_b_quark_unc)
        set_hist_error(higher_c_quark,higher_c_quark_unc)
        set_hist_error(higher_quark,higher_quark_unc)
        set_hist_error(higher_gluon,higher_gluon_unc)            
            
        if particle == "Quark":
            higher_hist_array.append(higher_quark)
            lower_hist_array.append(lower_quark)
    
        if particle == "Gluon":
            higher_hist_array.append(higher_gluon)
            lower_hist_array.append(lower_gluon)
    
        if particle == "B Quark":
            higher_hist_array.append(higher_b_quark)    
            lower_hist_array.append(lower_b_quark)    
    
        if particle == "C Quark":
            higher_hist_array.append(higher_c_quark)
            lower_hist_array.append(lower_c_quark)
                
    c = TCanvas("","",800,800)
        
    gStyle.SetOptStat(0)
    gPad.SetTickx()
    gPad.SetTicky()
    """
    gStyle.SetCanvasColor(-1)
    gStyle.SetPadColor(-1)
    gStyle.SetFrameFillColor(-1)
    gStyle.SetHistFillColor(-1)
    gStyle.SetTitleFillColor(-1)
    gStyle.SetFillColor(-1)
    gStyle.SetFillStyle(4000)
    gStyle.SetStatStyle(0)
    gStyle.SetTitleStyle(0)
    gStyle.SetCanvasBorderSize(0)
    gStyle.SetFrameBorderSize(0)
    gStyle.SetLegendBorderSize(0)
    gStyle.SetStatBorderSize(0)
    gStyle.SetTitleBorderSize(0)
    """
                
    print(particle)
    higher_hist_array[0].GetXaxis().SetTitle('pT(GeV)')
    higher_hist_array[0].GetYaxis().SetTitle(particle + ' Fraction')
    higher_hist_array[0].SetMaximum(1)
    higher_hist_array[0].SetMinimum(0)

    higher_hist_array[0].SetMarkerStyle(8)
    higher_hist_array[1].SetMarkerStyle(2)
    higher_hist_array[2].SetMarkerStyle(3)
    higher_hist_array[3].SetMarkerStyle(4)
    higher_hist_array[4].SetMarkerStyle(5)
    higher_hist_array[5].SetMarkerStyle(6)  


    higher_hist_array[0].SetMarkerSize(1.2)
    higher_hist_array[1].SetMarkerSize(1.2)
    higher_hist_array[2].SetMarkerSize(1.2)
    higher_hist_array[3].SetMarkerSize(1.2)
    higher_hist_array[4].SetMarkerSize(1.2)
    higher_hist_array[5].SetMarkerSize(1.2)

    leg1 = TLegend(.65,0.75,0.83,0.88)

    for mc_index in range(len(f_mc_generators)):

        higher_hist_array[mc_index].GetXaxis().SetRangeUser(500.,2000.)
        leg1.AddEntry(higher_hist_array[mc_index],mc_generators_name[mc_index],"p")
        if mc_index == 0:
            higher_hist_array[mc_index].Draw("L P0 E2")
        else:
            higher_hist_array[mc_index].Draw("L P0 E2 same")


    leg1.Draw()
    myText(0.15,0.84,'#it{#bf{#scale[1.4]{#bf{ATLAS} Internal}}}')
    myText(0.15,0.81,"#bf{#scale[1.2]{higher \eta jet}}")


    c.Print(particle + "_higher_fraction.pdf")

    c1 = TCanvas("","",800,800)

    lower_hist_array[0].GetXaxis().SetTitle('pT(GeV)')
    lower_hist_array[0].GetYaxis().SetTitle(particle + ' Fraction')

    lower_hist_array[0].SetMaximum(1)
    lower_hist_array[0].SetMinimum(0)

    lower_hist_array[0].SetMarkerStyle(8)
    lower_hist_array[1].SetMarkerStyle(2)
    lower_hist_array[2].SetMarkerStyle(3)
    lower_hist_array[3].SetMarkerStyle(4)
    lower_hist_array[4].SetMarkerStyle(5)
    lower_hist_array[5].SetMarkerStyle(6)  


    lower_hist_array[0].SetMarkerSize(1.2)
    lower_hist_array[1].SetMarkerSize(1.2)
    lower_hist_array[2].SetMarkerSize(1.2)
    lower_hist_array[3].SetMarkerSize(1.2)
    lower_hist_array[4].SetMarkerSize(1.2)
    lower_hist_array[5].SetMarkerSize(1.2)

    leg1 = TLegend(.65,0.75,0.83,0.88)

    for mc_index in range(len(f_mc_generators)):
        lower_hist_array[mc_index].GetXaxis().SetRangeUser(500.,2000.)
        leg1.AddEntry(lower_hist_array[mc_index],mc_generators_name[mc_index],"p")
        if mc_index == 0:
            lower_hist_array[mc_index].Draw("L P0 E2")
        else:
            lower_hist_array[mc_index].Draw("L P0 E2 same")

    leg1.Draw()
    myText(0.15,0.84,'#it{#bf{#scale[1.4]{#bf{ATLAS} Internal}}}')
    myText(0.15,0.81,"#bf{#scale[1.2]{lower \eta jet}}")


    c1.Print(particle + "_lower_fraction.pdf")
```
